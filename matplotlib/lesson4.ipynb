### TASK 2

```python
t = np.linspace(0, 10, num=51)
f = np.array([np.cos(_) for _ in t])
plt.plot(t, f)
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

### TASK 3

```python
x = np.linspace(-3, 3, num=51)
y1 = [_**2 for _ in x]
y2 = [2*_+0.5 for _ in x]
y3 = [-3*_ - 1.5 for _ in x]
y4 = [np.sin(_) for _ in x]
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax1.set_title('График 1')
ax2.plot(x, y2)
ax2.set_title('График 2')
ax3.plot(x, y3)
ax3.set_title('График 3')
ax4.plot(x, y4)
ax4.set_title('График 4')
fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()
```
